```python
im_orig = imread('/home/giorgia/Desktop/MAI/Thesis/images/DJI_0844.JPG')
im = rgb2lab(im_orig)
plt.figure(figsize=(20,30))
plt.subplot(131)
plt.imshow(im[:,:,0])
plt.subplot(132)
plt.imshow(im[:,:,1])
plt.subplot(133)
plt.imshow(im[:,:,2])
```

```python
from skimage.filters import try_all_threshold

try_all_threshold(im[:,:,0]);
```

```python
from skimage.filters import threshold_otsu, threshold_triangle

im_th_otsu = im[:,:,0]>threshold_otsu(im[:,:,0])
im_th_triangle = im[:,:,0]>threshold_triangle(im[:,:,0])
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(im_th_otsu, cmap ='gray')
plt.subplot(122)
plt.imshow(im_th_triangle, cmap ='gray')
```

```python
from skimage.morphology import disk, square, binary_dilation, remove_small_objects

im_filtered = remove_small_objects(im_th_triangle + im_th_otsu, 300)
im_filled = binary_dilation(im_filtered, selem=disk(5))
plt.figure(figsize=(10,10))
plt.imshow(im_filled, cmap='gray')
```
